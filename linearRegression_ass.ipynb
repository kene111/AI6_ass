```python
data3 = pd.read_csv("re_modelled.csv", sep=',', header = None)
data3.columns=["Annual barely yields","Per Acre"]
```

```python
data3 = data3.values
```

```python
x = np.c_[data3[:,0], np.ones(data3.shape[0])] 

print(x)
```

```python
y = np.c_[data3[:,1]]
```

```python
plt.scatter(x[:,0], y, s=30, marker='o')
```

```python
iterations = 2500
np.zeros(iterations)
```

```python
def train(x, y):
    iteration = 2500
    
    theta = [[0],[0]]
 
    cost_function = np.zeros(iteration)
    
    for i in range(iteration):
     
        h = x @ theta
    
        theta = update_weight(h, theta, x)
        
        cost_function[i] = cost(x, theta)

        print("The cost function is ", cost_function[i])
        
    return theta, cost_function
```

```python
def cost(x, theta):
    m = x[:,0].size
    h = x @ theta
    return (1/2*m) * np.sum(np.square(h - y))
```

```python
def update_weight(h, theta, x):
    m = x[:,0].size
    alpha = 0.01
    theta = theta - alpha * (1/m * (x.T @ (h - y)))
    return theta
```

```python
theta, cost_values = train(x, y)
```

```python
cost_values
```

```python
print('Our learned value of theta: ',theta.ravel())
```

```python
plt.plot(cost_values)
plt.ylabel('Cost J')
plt.xlabel('Iterations');
```

```python
x_plot = np.arange(1890,1940)
y_plot = theta[0]*x_plot + theta[1]
# Plot gradient descent
plt.scatter(x[:,0], y, s=30, c='r', marker='o', linewidths=1)
plt.plot(x_plot,y_plot, label='Linear regression (Gradient descent)') 
```
